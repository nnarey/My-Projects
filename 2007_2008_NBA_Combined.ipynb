## Getting Data

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.basketball-reference.com/leagues/NBA_2008_advanced.html"

def scrape_basketball_reference(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        table = soup.find('table')
        df = pd.read_html(str(table))[0]
        return df
    else: # Print an error message if the request was not successful
        print('Error: Unable to retrieve data from the provided URL')
        return None

df_adv = scrape_basketball_reference(url)
df_adv = df_adv.drop_duplicates(subset=['Player'], keep='first')
df_adv
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.basketball-reference.com/leagues/NBA_2008_per_game.html"

def scrape_basketball_reference(url):
    # Send a GET request to the specified URL
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find the table containing the data you want to scrape
        table = soup.find('table')

        # Convert the HTML table to a pandas DataFrame
        df = pd.read_html(str(table))[0]

        return df
    else:
        # Print an error message if the request was not successful
        print('Error: Unable to retrieve data from the provided URL')
        return None

df_pg = scrape_basketball_reference(url)
df_pg = df_pg.drop_duplicates(subset=['Player'], keep='first')
df_pg
```

```python
result = df_adv.merge(df_pg, on=("Player", "Pos", "Age", "Tm", "G", "Rk"))
result
```

```python
result["PTS"] = pd.to_numeric(result["PTS"], errors='coerce', downcast='float', )
result = result[result["PTS"] >= 15.5]
result["TRB"] = pd.to_numeric(result["TRB"], errors='coerce', downcast='float', )
result = result[result["TRB"] >= 2.5]
result["AST"] = pd.to_numeric(result["AST"], errors='coerce', downcast='float', )
result = result[result["AST"] >= 2.5]
result["STL"] = pd.to_numeric(result["STL"], errors='coerce', downcast='float', )
result = result[result["STL"].astype(float) >= 0.5]
result["BLK"] = pd.to_numeric(result["BLK"], errors='coerce', downcast='float', )
result = result[result["BLK"].astype(float) >= 0.1]
result["BLK"] = pd.to_numeric(result["G"], errors='coerce', downcast='float', )
result = result[result["G"].astype(float) >= 47]
#result.set_index("Rk", inplace=True)
result["Player"] = result["Player"].str.replace("*", "")
result
```

```python
result.to_csv("2024.csv")
```
