# CONCACAF

### 2023 Gold Cup

```python
# CONCACAF GOLD CUP 2023
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://fbref.com/en/comps/681/stats/Gold-Cup-Stats"

def scrape_football_reference(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        table = soup.find('table')
        df = pd.read_html(str(table))[0]
        return df
    else: # Print an error message if the request was not successful
        print('Error: Unable to retrieve data from the provided URL')
        return None

df_CCF = scrape_football_reference(url)
df_CCF
```

```python
df_CCF.columns = range(df_CCF.shape[1])
df_CCF = df_CCF.T.reset_index().T.reset_index(drop=True).rename(columns=lambda x: f'Column_{x}')
df_CCF = df_CCF.iloc[1:]
```

```python
df_CCF = df_CCF.drop(columns=['Column_1', 'Column_5', 'Column_6', 'Column_7'])
```

```python
df_CCF
```

```python
df_CCF = df_CCF.rename(columns={'Column_0': 'Nation', 'Column_2': 'Average Age', 'Column_3': 'Possession %', 'Column_4': 'Matches Played',
                                'Column_8': 'Goals', 'Column_9': 'Assists', 'Column_10': 'G+A', 'Column_11': 'G-Pk', 'Column_12': 'Pk', 'Column_13': 'Pk Att',
                                'Column_14':'Yellow Cards', 'Column_15':'Red Cards', 'Column_16': 'G/90',
                                'Column_17': 'A/90', 'Column_18': 'G+A/90','Column_19':'G-Pk/90', 'Column_20':'G+A-Pk/90'})
df_CCF
```

```python
df_CCF['Nation'] = df_CCF['Nation'].str[2:]
df_CCF
```

```python
# Add in Rows For Points

df_CCF["GC Points"] = [6, 4, 0, 2, 4, 7, 3, 4, 10, 3, 15, 11, 4, 0, 3, 9]
df_CCF = df_CCF.sort_values(by='GC Points', ascending=False)
df_CCF
```

```python
df_CCF["NL Points"] = [9, 6, 6, 15, 0, 3, 0, 0, 3, 0, 0, 0, 3, 0, 0, 0]
df_CCF["COPA Points"] = [4, 6, 0, 3, 0, 5, 4, 0, 0, 0, 0, 0, 0, 0, 0, 0]
df_CCF["WC Points"] = [4, 0, 0, 5, 0, 0, 3, 0, 0, 0, 0, 0, 0, 0, 0, 0]
df_CCF["WC Points"] = df_CCF["WC Points"] * 4
df_CCF["T2 Points"] = df_CCF["COPA Points"] * 3
df_CCF["Total Points"] = 0.5 * (df_CCF["GC Points"] + df_CCF["NL Points"]) + df_CCF["COPA Points"] + df_CCF["WC Points"]
df_CCF = df_CCF.sort_values(by='Total Points', ascending=False)
df_CCF["T3 Points"] = (df_CCF["GC Points"] + df_CCF["NL Points"]) * 0.5
df_CCF["Total Points"] = df_CCF["WC Points"] + df_CCF["T2 Points"] + df_CCF["T3 Points"]
df_CCF = df_CCF[["Nation", "WC Points", "T2 Points", "T3 Points", "Total Points"]]
df_CCF
```

# UEFA

```python
url = "https://fbref.com/en/comps/676/stats/UEFA-Euro-Stats"
df_UEFA = scrape_football_reference(url)
df_UEFA
```

```python
df_UEFA.columns = range(df_UEFA.shape[1])
df_UEFA = df_UEFA.T.reset_index().T.reset_index(drop=True).rename(columns=lambda x: f'Column_{x}')
df_UEFA = df_UEFA.iloc[1:]
df_UEFA
```

```python
df_UEFA = df_UEFA.drop(columns=['Column_1', 'Column_5', 'Column_6', 'Column_7'])
df_UEFA = df_UEFA.rename(columns={'Column_0': 'Nation', 'Column_2': 'Average Age', 'Column_3': 'Possession %', 'Column_4': 'Matches Played',
                                'Column_8': 'Goals', 'Column_9': 'Assists', 'Column_10': 'G+A', 'Column_11': 'G-Pk', 'Column_12': 'Pk', 'Column_13': 'Pk Att',
                                'Column_14':'Yellow Cards', 'Column_15':'Red Cards', 'Column_16': 'G/90',
                                'Column_17': 'A/90', 'Column_18': 'G+A/90','Column_19':'G-Pk/90', 'Column_20':'G+A-Pk/90'})
df_UEFA
```

```python
df_UEFA['Nation'] = df_UEFA['Nation'].str[3:]
df_UEFA['T2 Points'] = [1, 6, 4, 2, 1, 3, 10, 9, 4, 11, 3, 4, 10, 1, 8, 4, 1, 2, 5, 4, 19, 9, 9, 4]
df_UEFA['T2 Points'] = df_UEFA['T2 Points'] * 3
df_UEFA["WC Points"] = [0, 0, 4, 7, 0, 1, 10, 16, 0, 4, 0, 0, 11, 4, 9, 0, 0, 1, 0, 0, 5, 6, 0, 0]
df_UEFA['WC Points'] = df_UEFA['WC Points'] * 4
df_UEFA["T3 Points"] = [0, 4, 10, 17, 4, 12, 3, 5, 0, 7, 10, 11, 16, 7, 10, 0, 0, 0, 0, 0, 15, 9, 0, 0]
df_UEFA["T3 Points"] = df_UEFA["T3 Points"] * 2
df_UEFA['Total Points'] = (df_UEFA['T2 Points']) + (df_UEFA['WC Points']) + df_UEFA['T3 Points']
df_UEFA = df_UEFA.sort_values(by='Total Points', ascending=False)
df_UEFA = df_UEFA[['Nation', 'WC Points', 'T2 Points', 'T3 Points', 'Total Points']]
df_UEFA
```

```python
merged_df = pd.concat([df_CCF, df_UEFA], ignore_index=True)
merged_df = merged_df.sort_values(by='Total Points', ascending=False)
merged_df['Rank'] = merged_df['Total Points'].rank(method='dense', ascending=False)
merged_df["Rank"] = merged_df["Rank"].astype(int)
merged_df = merged_df.set_index('Rank')
merged_df
```

Games from over four years ago should have minimal impact on the ranking status of each team.

This ranking system also gives equal opportunity for nations that perform at a high level in non Europe and South American confederations to compete in their tournaments and make a name for themselves.
